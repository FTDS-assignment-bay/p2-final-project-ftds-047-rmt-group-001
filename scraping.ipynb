# **Import Libraries**

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
```

# **Scraping**

## **SUN PROTECTION**

```python
# SETUP

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
wait = WebDriverWait(driver, 10)

url_category = "https://reviews.femaledaily.com/products/moisturizer/sun-protection-1"
driver.get(url_category)
time.sleep(3)


# LOAD MORE PRODUCTS

MAX_LOAD = 10
for i in range(MAX_LOAD):
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        btn = WebDriverWait(driver, 6).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "button.btn-load-more"))
        )
        driver.execute_script("arguments[0].scrollIntoView();", btn)
        time.sleep(1)
        btn.click()
        print(f"Load more clicked {i+1}")
        time.sleep(3)
    except Exception:
        print("No more Load More button found or not clickable")
        break


# EXTRACT PRODUCT LIST

products = driver.find_elements(By.CSS_SELECTOR, "a.product-card")
results = []

for prod in products:
    try:
        brand = prod.find_element(By.CSS_SELECTOR, "p.text-ellipsis.word-break").text
    except:
        brand = None

    try:
        name = prod.find_element(By.CSS_SELECTOR, "p.two-line.word-break").text
    except:
        name = None

    try:
        rating_text = prod.find_element(By.CSS_SELECTOR, "div.rating").text.strip()
        rating_text = rating_text.replace("\n", " ")
        rating = rating_text.split(" ")[0]
        review_count = rating_text.split("(")[1].replace(")", "")
    except:
        rating = None
        review_count = None

    try:
        url = prod.get_attribute("href")
    except:
        url = None

    results.append({
        "brand": brand,
        "product": name,
        "rating": rating,
        "review_count": review_count,
        "url": url
    })

print(f"\n Total products scraped: {len(results)}")


# SCRAPE PRODUCT DETAILS (img, price, desc)

details = []
for i, prod in enumerate(results, start=1):
    product_url = prod["url"]
    if not product_url:
        continue

    print(f"\n[{i}] {product_url}")
    driver.get(product_url)
    time.sleep(2)

    # Scroll to ensure elements loaded
    for _ in range(15):
        driver.execute_script("window.scrollBy(0, 300);")
        time.sleep(0.2)

    # Image
    try:
        img_el = driver.find_element(By.CSS_SELECTOR, "div.image-wrapper img")
        img_url = img_el.get_attribute("src")
        print("Image found")
    except:
        img_url = None
        print("Image not found")

    # Price
    try:
        price_el = driver.find_element(By.CSS_SELECTOR, "p.product-price")
        price = price_el.text.strip()
        print("Price found:", price)
    except:
        price = None
        print("Price not found")

    # Expand Description
    try:
        expand_btn = wait.until(
            EC.presence_of_element_located((By.ID, "id_button_description"))
        )
        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", expand_btn)
        time.sleep(1)
        expand_btn.click()
        print("Expand button clicked")
        time.sleep(1.2)
    except:
        print("Expand button not clickable")

    # Description
    try:
        desc_el = wait.until(
            EC.visibility_of_element_located((By.CSS_SELECTOR, "p[class*='product-desc']"))
        )
        desc = desc_el.text.strip()
        print("Description found")
    except:
        desc = None
        print("Description not found")

    prod["image_url"] = img_url
    prod["price"] = price
    prod["description"] = desc
    details.append(prod)


# SAVE RESULT

driver.quit()

final_df = pd.DataFrame(details)
final_df["product_type"] = "sun protection"
cols = ["product_type"] + [col for col in final_df.columns if col != "product_type"]
final_df = final_df[cols]

final_df.to_csv("sun_protection_full.csv", index=False)
print("\nScraping complete!")
print(final_df.head())
```
