## 5 - Handle Missing Value

### **Drop Missing Value**

#### **Kolom `product` dan `description` dengan missing value**

```python
df = df[df['product'].notna() & df['description'].notna()].copy().reset_index(drop=True)
```

#### **Kolom `description` dengan value `There's no description for this product.`**

```python
df = df.drop(df[df["description"] == "There's no description for this product."].index)
```

```python
df.info()
```

*Informasi :*
- Baris pada kolom `product` dan `description` dengan missing value dilakukan hapus kolom karena karena kedua kolom tersebut merupakan informasi utama produk.
- Nilai kosong pada kolom ini membuat data tidak dapat digunakan untuk analisis deskripsi maupun sistem rekomendasi, sehingga baris tersebut dianggap tidak valid dan dihapus dari dataset.
- Kolom `description` dengan nilai `There's no description for this product.` akan dilakukan drop baris karena tidak bisa dilakukan modelling dengan value tersebut.
- Missing value pada kolom `price` akan dilakukan imputasi dengan harga manual yang ada pada platform shopee, namun sebelum itu akan dilakukan perubahan tipe data pada kolom `price` tersebut.

## 6 - Ubah Tipe Kolom

#### **Ubah tipe kolom `review`**

```python
df['review_count'] = df['review_count'].astype(int)
```

#### **Ubah tipe kolom `price`**

```python
# Bersihkan format harga
df['price'] = (
    df['price']                       
    .str.replace(r'Rp\.?', '', regex=True)  
    .str.replace(r'\.', '', regex=True)    
    .str.replace(r'\s+', '', regex=True)   
    .replace('', np.nan)                      
    .astype(float)                         
)
df['price']
```

## 8 - Membuat Unique Column

```python
# Buat kolom unique_id dari brand + product
df['unique_id'] = (
    df['brand'].fillna('').str.strip() + ' - ' + df['product'].fillna('').str.strip()
)

# Tambahkan index hanya untuk yang duplikat
dupes = df['unique_id'].duplicated(keep=False)
df.loc[dupes, 'unique_id'] = (
    df.loc[dupes, 'unique_id'] + " - " + df.loc[dupes].index.astype(str)
    )
```
